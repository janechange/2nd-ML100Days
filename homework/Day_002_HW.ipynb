#### 截取部分資料

Locate row 1000

```python
app_train.iloc[1000]
```

Check all genders

```python
gd = app_train[["SK_ID_CURR", "CODE_GENDER"]]
gd
```

Locate column 20~25, rows 100~110

```python
df=[]
for i in range(20, 25):
    df.append(app_train[app_train.columns[i]][100:110])
res = pd.concat(df, axis = 1, join = 'outer')
print(res)
```

Fill NAN

```python
res1 = res.fillna(0)
res1
```

Drop NAN

```python
res2 = res.dropna()
res2
```
